Необходимо привести названия столбцов к нижнему регистру и заменить пробелы на знак нижнего подчеркивания

```python
import pandas as pd
import numpy as np
```

```python
bookings = pd.read_csv('2_bookings.csv', sep=';')
```

```python
# первый вариант
bookings.columns = bookings.columns.str.replace(pat=' ', repl='_').str.lower()
```

```python
# второй вариант
rename_cols = {}
for col in bookings.columns:
    rename_cols[col] = col.replace(' ', '_').lower()
    
bookings=bookings.rename(columns=rename_cols)
```
